```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib notebook
%matplotlib inline
```

```python
veri1_ort = []
for i in range(50) :
    veri1 = np.random.normal(size=1000) # ortalalama ve ss'yi secmedim.
    veri1_ort.append(np.mean(veri1))
veri2_ort = []
for ii in range(1000) :
    veri2 = np.random.normal(size=1000)
    veri2_ort.append(np.mean(veri2))
plt.figure(figsize=(16,6))
plt.subplot(121)
plt.hist(veri1_ort, bins=33, label='50 Kez') # ortalamalar normal dagilima uymuyor.
plt.subplot(122)
plt.hist(veri2_ort, bins=33, label='1000 Kez') # ortalamalar normal dagilima uyuyor.
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
```

```python
hayvan_kilo = pd.DataFrame()
hayvan_kilo['Fil'] = np.append(np.random.normal(4000,322,33), np.random.normal(4700,250,33))
hayvan_kilo['Aslan'] = np.append(np.random.normal(188,21,33), np.random.normal(200,33,33))
hayvan_kilo['Antilop'] = np.append(np.random.normal(222,22,33), np.random.normal(313,44,33))
hayvan_kilo
```

```python
hayvan_kilo.hist(bins=22)
```

```python
# built in fonksiyonlarla 'merkezi egilim' hesaplama :
  # ortalama :
print(hayvan_kilo.Aslan.mean())
print(hayvan_kilo.Antilop.mean())
print(hayvan_kilo.Fil.mean(), '\n')

# median:
print(hayvan_kilo.Aslan.median())
print(hayvan_kilo.Antilop.median())
print(hayvan_kilo.Fil.median())

# mod : (hicbir deger birden fazla tekrar etmemis. dolayisiyla modu yok. hesaplanamiyor)
#hayvan_kilo.Aslan.mode()
#hayvan_kilo.Fil.mode()
#hayvan_kilo.Antilop.mode()
```

```python
# matematiksel hesaplamalarla 'merkezi egilim' olcumleri :
  # ortalama :
print(sum(hayvan_kilo.Aslan)/len(hayvan_kilo.Aslan))
print(sum(hayvan_kilo.Antilop)/len(hayvan_kilo.Antilop))
print(sum(hayvan_kilo.Fil)/len(hayvan_kilo.Fil), '\n')

  # median :
asln=np.sort(hayvan_kilo.Aslan)
antlp=np.sort(hayvan_kilo.Antilop)
fl=np.sort(hayvan_kilo.Fil)
print((asln[32]+asln[33])/2) # eleman sayisi cift (66) oldugu icin ortadaki [32][33] iki eleanin aritmetik ortalamasi alinir
print((antlp[32]+antlp[33])/2)
print((fl[32]+fl[33])/2)
```

```python
# built in ve kutuphanelerle 'dagilim' ilcumleri :
  # varyans :
print(hayvan_kilo.Aslan.var())
print(hayvan_kilo.Antilop.var())
print(hayvan_kilo.Fil.var(), '\n')

  # standart sapma :
print(hayvan_kilo.Aslan.std())
print(hayvan_kilo.Antilop.std())
print(hayvan_kilo.Fil.std(), '\n')

  # standart hata :
print((hayvan_kilo.Aslan.std())/np.sqrt(len(hayvan_kilo.Aslan)))
print((hayvan_kilo.Antilop.std())/np.sqrt(len(hayvan_kilo.Antilop)))
print((hayvan_kilo.Fil.std())/np.sqrt(len(hayvan_kilo.Fil)))
```

```python
# matematiksel yontemlerle 'dagilim' olcumleri :
  # varyanas :
def variant(arr) :
    var_x = []
    for i in arr :
        var_x.append((i-arr.mean())**2)
    return sum(var_x)/(len(arr)-1)

print(variant(hayvan_kilo.Aslan))
print(variant(hayvan_kilo.Antilop))
print(variant(hayvan_kilo.Fil), '\n')

  # standart sapma :
print((variant(hayvan_kilo.Aslan))**0.5)
print((variant(hayvan_kilo.Antilop))**0.5)
print((variant(hayvan_kilo.Fil))**0.5, '\n')

  # standart hata : onceki ile ayni
```
